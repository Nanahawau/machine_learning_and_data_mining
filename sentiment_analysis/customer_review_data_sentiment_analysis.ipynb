### Preprocessing

```python
dataset_with_good_columns = dataset.drop(columns=['Unnamed: 0', 'Clothing ID', 'Recommended IND'])
dataset_with_good_columns.head()
```

```python

from transformers import BertTokenizer
import tensorflow_hub as hub 

# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Load BERT model from TensorFlow Hub
model = hub.load("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/3")

# Function to generate BERT embeddings using TensorFlow
def get_bert_embeddings_tf(text):
    tokens = tokenizer(text, return_tensors='tf', padding=True, truncation=True, max_length=512)
    outputs = model(tokens)
    # Use the CLS token embedding as the representation
    return outputs.last_hidden_state[:, 0, :].numpy()
```

```python


import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

# download required resources (run once)
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
nltk.download("averaged_perceptron_tagger")
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')

# helpers
_stopwords = set(stopwords.words("english"))
_ps = PorterStemmer()
_wl = WordNetLemmatizer()

# map NLTK POS tags to wordnet POS for better lemmatization
from nltk.corpus import wordnet
from nltk import pos_tag

def _pos_to_wordnet(tag):
    tag = tag[0].upper()
    return {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}.get(tag, wordnet.NOUN)

def clean_text(text, remove_stopwords=True, do_stem=False, do_lemma=True):
    if not isinstance(text, str):
        text = "" if text is None else str(text)
    text = text.lower()
    text = re.sub(r"http\S+|www\.\S+", " ", text)        # remove URLs
    text = re.sub(r"<[^>]+>", " ", text)                # remove HTML
    text = re.sub(r"[^a-z0-9\s']", " ", text)           # keep alnum + apostrophe
    tokens = word_tokenize(text)
    if remove_stopwords:
        tokens = [t for t in tokens if t not in _stopwords and len(t) > 1]
    if do_lemma:
        pos_tags = pos_tag(tokens)
        tokens = [_wl.lemmatize(tok, _pos_to_wordnet(tag)) for tok, tag in pos_tags]
    if do_stem:
        tokens = [_ps.stem(tok) for tok in tokens]
    return " ".join(tokens)

# if dataset has both review and title, create a combined field
dataset_with_good_columns["title"] = dataset_with_good_columns.get("title", dataset_with_good_columns.get("Title", None))
dataset_with_good_columns["review_text"] = dataset_with_good_columns.get("review text", dataset_with_good_columns.get("Review Text", None))

# safe concat (handles missing)
dataset_with_good_columns["text_combined"] = (
    dataset_with_good_columns["title"].fillna("").astype(str).str.strip() + " " +
    dataset_with_good_columns["review_text"].fillna("").astype(str).str.strip()
).str.strip()

# then use "text_combined" for cleaning/vectorizing
dataset_with_good_columns["clean_text"] = dataset_with_good_columns["text_combined"].map(lambda s: clean_text(s, remove_stopwords=True, do_stem=True, do_lemma=False))


print("Text column used:", "text combined")
display(dataset_with_good_columns[["text_combined", "clean_text"]].head(6))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all cleaned text into a single string
text_data = " ".join(dataset_with_good_columns["clean_text"].dropna())

# Generate the word cloud
wordcloud = WordCloud(
    width=800, 
    height=400, 
    background_color="white", 
    colormap="viridis", 
    max_words=200
).generate(text_data)

# Visualize the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Cleaned Text Data", fontsize=16)
plt.show()
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.pipeline import Pipeline
import tensorflow_hub as hub
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.utils.class_weight import compute_class_weight
from sklearn.feature_extraction.text import TfidfVectorizer


# Prepare features
# categorical_cols = ["Department Name", "Division Name", "Class Name"]
# numerical_cols = ["Age", "Positive Feedback Count"]
target_col = "Rating"  
y = dataset_with_good_columns[target_col] - 1

# Convert text to TF-IDF representation
vectorizer = TfidfVectorizer() 
X = vectorizer.fit_transform(dataset_with_good_columns["clean_text"])


# Printing  Vectorization Features

feature_names = vectorizer.get_feature_names_out()
print("Feature Names:", feature_names)

# Convert the sparse matrix to a DataFrame for easier visualization
vectorized_output = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Display the output
print(vectorized_output)

# Split the dataset into training and test sets first
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
      test_size=0.2, stratify=y, random_state=42
)


# Apply SMOTE to the training set
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)



print("Class distribution after SMOTEENN:")
print(Counter(y_train_resampled))

# Check class distribution before SMOTE
# Before SMOTE
plt.figure(figsize=(8, 4))
y_train.value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Class Distribution Before SMOTE")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# After SMOTE
plt.figure(figsize=(8, 4))
y_train_resampled.value_counts().plot(kind='bar', color='salmon', edgecolor='black')
plt.title("Class Distribution After SMOTE")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()


# Train a model (e.g., Random Forest)
clf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight="balanced")
clf.fit(X_train_resampled, y_train_resampled)

# Evaluate the model
y_pred = clf.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix_rf = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()



```
